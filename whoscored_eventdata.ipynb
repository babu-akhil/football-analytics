```python
import soccerdata as sd
```

```python
ws = sd.WhoScored(leagues="ENG-Premier League", seasons=2024)
print(ws.__doc__)
```

```python
epl_schedule = ws.read_schedule()
epl_schedule.head()
```

```python
events = ws.read_events(match_id=1821049)
events.head()
```

```python
events.type
events_pass = events[events['type'] == 'Pass']
events_pass

successful_passes = events_pass[events_pass['outcome_type'] == 'Successful']
successful_passes

events_pass_team = events_pass.groupby('team')['player'].count()
print(events_pass_team)

successful_passes_team = successful_passes.groupby('team')['player'].count()
print(successful_passes_team)

passing_accuracy = (successful_passes_team / events_pass_team) * 100
passing_accuracy = passing_accuracy.round(2)
print(passing_accuracy)
```
